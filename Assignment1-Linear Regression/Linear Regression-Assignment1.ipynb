```python
'''Problem Statement:
You work in XYZ Company as a Python Data Scientist. The company officials
have collected some data on salaries based on year of experience and wish for
you to create a model from it.
Dataset: data.csv
Tasks To Be Performed:
1. Load the dataset using pandas
2. Extract data from years experience column is a variable named X
3. Extract data from the salary column is a variable named Y
4. Divide the dataset into two parts for training and testing in 66% and 33%
proportion
5. Create and train Linear Regression Model on training set
6. Make predictions based on the testing set using the trained model
7. Check the performance by calculating the r2 score of the model'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#importing matplotlib,numpy and pandas libraries
```

```python
#1.Load the dataset using pandas
#importing the csv file to DataFrame using pandas

df=pd.read_csv("data.csv")
df
```

```python
df.info()

#checking the column,datatype and if there are any non-null values
```

```python
#2. Extract data from years experience column is a variable named X
#This is the feature-independent variable

X=df.Salary.values
print(X)
```

```python
# 3. Extract data from the salary column is a variable named Y
#This is the target-dependent variable

Y=df.YearsExperience.values
print(Y)
```

```python
#4. Divide the dataset into two parts for training and testing in 66% and 33% proportion

#importing library to split the dataset
from sklearn.model_selection import train_test_split

#X_train,Y_train-training set
#X_test,Y_test-testing set
#test_size-denotes the proportion of testing set in the whole dataset
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.33)
```

```python
#5. Create and train Linear Regression Model on training set

#importing the library
from sklearn.linear_model import LinearRegression

#creating object of LinearRegression
lr=LinearRegression()

#convert to 2D with 1 column in each row so that model can use it for training
X_train=X_train.reshape(-1,1)

#passing training set for training
lr.fit(X_train,Y_train)
```

```python
#6. Make predictions based on the testing set using the trained model

#convert to 2D with 1 column in each row so that model can use it for testing
x_test=x_test.reshape(-1,1)

#predicting y_value using test set features
y_pred=lr.predict(x_test)
```

```python
#7. Check the performance by calculating the r2 score of the model

#import library
from sklearn.metrics import r2_score

#comparing the predicted y values and y_test values to find r2
r2=r2_score(y_test,y_pred)
print(r2)

#As r2 gets closer to 1, model gets more fit
```
